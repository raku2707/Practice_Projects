### Now we convert grades into the some numerical numbers.

```python
grades_enc = {'B-':2.7, 'A':4, 'D':1.0, 'A-':3.7, 'B+':3.4, 'B':3.0, 'C+':2.4, 'C':2.0, 'C-':1.7, 'D+':1.4, 'WU':0.0, 'A+':4,'F':0.0,'W':0,'I':0}
```

### Replace number from grades

```python
for item in input_var:
      input_var[item] = input_var[item].replace(grades_enc)

input_var.head()
```

```python
# Concating inputs and targets for data insights
SG_pred = pd.concat([input_var, target_var], axis=1)
```

## Splitting file into differnt columns

```python
#Splitting file into different columns based on our model prediction requirement
model1_columns = []
model2_columns = []
model3_columns = []

for item in [input_var]:
    for i in item:
        if i[3]=='1':
            model1_columns.append(i)
            
        elif i[3]=='2':
            model2_columns.append(i)
        
        else:
            model3_columns.append(i)

model2_columns = model1_columns + model2_columns
model3_columns = model2_columns + model3_columns
```

## Prepare the data for Linear Regression:

### Model 1: Predict final CGPA based on grade predictions of first year only.

```python
# Input data
labels = SG_pred['CGPA'].values
features = SG_pred[list(model1_columns)].values
```

##### Splitting data into train_test_split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.30)
```

```python
#Import library
from sklearn import linear_model
```

```python
#Bulid a new Model

lr1 = linear_model.LinearRegression()
```

```python
# Train the Model

lr1.fit(X_train, y_train)
```

```python
#Predict for training set

y_train_predict = lr1.predict(X_train)
```

```python
# Predict for testing set

y_test_predict = lr1.predict(X_test)
```

```python
input_lr=[]

input_lr.append([4, 3.0, 0, 0, 0, 1.0, 4, 2.0, 0, 4, 4])
print(input_lr )
print(lr1.predict(input_lr ))
```

```python
#Import library

from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
#model evaluation for training set

mse_train = mean_squared_error(y_train, y_train_predict)
mae_train = mean_absolute_error(y_train, y_train_predict)
print("The model performance for training set")
print('MSE is {}'.format(np.sqrt(mse_train)))
print('MAE is {}'.format(mae_train))
```

```python
# model evaluation for testing set

mse_test = mean_squared_error(y_test, y_test_predict) # Return the loss it is non-negative . The best value is 0.0
mae_test = mean_absolute_error(y_test, y_test_predict) # Return the loss it is non-negative . The best value is 0.0
print("The model performance for testing set")
print('MSE is {}'.format(np.sqrt(mse_test)))
print('MAE is {}'.format(mae_test))
```

```python
test_score= lr1.score(X_test, y_test)# Return R2 score . The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse) 
print("The score of model for testing set")
print('Test score {}'.format(test_score.round(2)))
```

```python
#Checking the actual and predict value

compare={"Actual": y_test, "Predict": y_test_predict}
pd.DataFrame(compare)
```

```python
#Visualize the Result

idx= np.random.choice(X_test.shape[0])
print ("Predicted CGPA :", y_test_predict[idx])
print ("Actual CGPA :", y_test[idx])
```

```python
# visualization

plt.figure(figsize=(22,5))
plt.scatter(y_test, y_test_predict,  color='black')
plt.scatter(y_test[idx], y_test_predict[idx])
plt.xlabel('CGPA')
plt.ylabel('Predicted CGPA)')
plt.plot([SG_pred["CGPA"].min(), SG_pred["CGPA"].max()], [SG_pred["CGPA"].min(), SG_pred["CGPA"].max()], color='red')
plt.show()
```

```python
#Visualize the scatterplot from model1

plt.figure(figsize=(22,5))
plt.scatter(y_test, y_test_predict,  color='blue')
plt.scatter(y_train, y_train_predict,  color='Orange')
plt.plot([SG_pred['CGPA'].min(), SG_pred['CGPA'].max()], [SG_pred['CGPA'].min(), SG_pred['CGPA'].max()], color='red')
plt.xlabel('CGPA')
plt.ylabel('Predicted CGPA')
plt.show()
```

## Prepare the data for Random Forest

### Model 1: Predict final CGPA based on grade predictions of first year only.

```python
# Input data
labels = SG_pred['CGPA'].values
features = SG_pred[list(model1_columns)].values
```

```python
#Splitting data into train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=0)
```

```python
#Import library

from sklearn.ensemble import RandomForestRegressor


#FItting the model in randomforestregressor

RFR_1 = RandomForestRegressor(n_estimators=20, random_state=0)
RFR_1.fit(X_train, y_train)
y_pred = RFR_1.predict(X_test)
```

```python
#Import library

from sklearn import metrics


#Cehcking the error value

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("The score of model for testing set",RFR_1.score(X_test, y_test))
```
