## Visualize distplot

```python
plt.figure(figsize=(20,25), facecolor='Green')
plotnumber=1

for column in Glass_ident:
    if plotnumber <=9:
        ax = plt.subplot(5,5, plotnumber)
        sns.distplot(Glass_ident[column], color='r')
        plt.xlabel(column, fontsize=15)
    plotnumber+=1
plt.show()
```

#### Outcomes

1. The distribution of potassium (K) and Barium (Ba) seem to contain many outliers.
2. The features Fe, Ba, Ca and K are Highly Skewed (They have High skew coefficients).
3. None of the features follow Normal Distribution (are not normally distributed).

## Checking outliers through Z Score

```python
from scipy import stats

z = abs(stats.zscore(Glass_ident))

#np.where(z > 3)

Glass_ident = Glass_ident[(z < 3).all(axis=1)]

#data.shape
```

```python
Glass_ident.shape
```

## Separating Features and Label

```python
#Separating the fratures and label

features = ['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']
label = ['glass_type']

X = Glass_ident[features]

y = Glass_ident[label]
```

```python
#SHape of X

X.shape
```

```python
#Type of X

type(X)
```

## Create train, test and split

```python
#Import train_test_split library
from sklearn.model_selection import train_test_split


#Create train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state=2)

print("x_train shape: ", x_train.shape)
print("x_test shape: ", x_test.shape)
print("y_train shape: ", y_train.shape)
print("x_test shape: ", y_test.shape)
```

# Applying Machine learning model

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression

Scores = []

for i in range(1):
    logistic = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',max_iter=100)
    logistic.fit(x_train, y_train)
    score = logistic.score(x_test,y_test)
    Scores.append(score)
    
print(logistic.score(x_train,y_train))
print(Scores)
```

## Desicion Tree

```python
from sklearn.tree import DecisionTreeClassifier

Scores = []

for i in range(1):
    tree = DecisionTreeClassifier(random_state=0)
    tree.fit(x_train, y_train)
    score = tree.score(x_test,y_test)
    Scores.append(score)

print(tree.score(x_train,y_train))
print(Scores)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

Scores = []

for i in range (2,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    score = knn.score(x_test,y_test)
    Scores.append(score)

print(knn.score(x_train,y_train))
print(Scores)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

Scores = []

for i in range (1):
    RFC = RandomForestClassifier(n_estimators=10, max_depth=2, random_state=0)
    RFC.fit(x_train, y_train)
    score = RFC.score(x_test,y_test)
    Scores.append(score)

print(RFC.score(x_train,y_train))
print(Scores)
```

## Support Vector Machine Classifier

```python
from sklearn.svm import SVC

Scores = []

for i in range(1):
    SVC = SVC(random_state=0)
    SVC.fit(x_train, y_train)
    score = SVC.score(x_test,y_test)
    Scores.append(score)

print(SVC.score(x_train,y_train))
print(Scores)
```

## Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier

Scores = []

for i in range(1):
    GBC = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
    GBC.fit(x_train, y_train)
    score = GBC.score(x_test,y_test)
    Scores.append(score)

print(GBC.score(x_train,y_train))
print(score)
```
